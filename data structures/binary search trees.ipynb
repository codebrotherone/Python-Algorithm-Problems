```python
"""
Binary search tree in python.

"""

class node:
    def __init__(self, value=None):
        self.value=value
        self.left=None
        self.right=None

class binary_search_tree:
    def __init__(self):
        self.root=None
    
    def insert(self, value):
        if not self.root:
            self.root = node(value)
        else:
            self._insert(value, self.root)
            
    def _insert(self, value, cur_node):
        if value < cur_node.value:
            if cur_node.left:
                self._insert(value,  cur_node.left)
            else:
                cur_node.left = node(value)
        elif value > cur_node.value:
            if cur_node.right:
                self._insert(value, cur_node.right)
            else:
                cur_node.right = node(value)
        else:
            print('Value already in tree')
    
    def display_tree(self, order='in'):
        print('{} order sort chosen'.format(order))
        if self.root:
            self._display_tree(self.root, order) 
        else:
            print('Nothing to display')
            
    def _display_tree(self, cur_node, order='in'):
        if cur_node:
            if order=='in':
                self._display_tree(cur_node.left)
                print(cur_node.value)
                self._display_tree(cur_node.right)
            elif order=='pre':
                print(cur_node.value)
                self._display_tree(cur_node.left)
                self._display_tree(cur_node.right)
            elif order=='post':
                self._display_tree(cur_node.left)
                self._display_tree(cur_node.right)
                print(cur_node.value)
```

```python
def make_tree(tree, num_elems = 100, max_int=1000):
    from random import randint
    for _ in range(num_elems):
        num = randint(0, max_int)
        tree.insert(num)
    return tree

t = binary_search_tree()
t = make_tree(t)

t.display_tree(order='in')
```
